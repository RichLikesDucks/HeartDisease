```python
def histogram(col_name, title, xlabel):
    fig, ax = plt.subplots()
    data[col_name].hist(grid=False)
    ax.set_title(title, fontsize=12)
    ax.set_xlabel(xlabel, fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)
```

```python
histogram('age', 'Patient Age Histogram', 'Age')
```

```python
histogram('chol', 'Cholesteral Histogram', 'Cholesteral')
```

```python
histogram('trestbps', 'Patient trestbps', 'trestbps')
```

```python
data_target = data[data['target'] == 1]

data_restbps = data[['trestbps','target']].groupby(['trestbps']).count()

data_restbps = data_restbps.rename(columns={'target': 'total_bps'})

data_restbps['total_heart'] = data_target[['trestbps','target']].groupby(['trestbps']).count()['target']

data_restbps['Heart Disease Per'] = data_restbps['total_heart']/data_restbps['total_bps']*100

data_restbps.fillna(0, inplace = True)

data_restbps = data_restbps['Heart Disease Per'].reset_index()

data_restbps
```

```python
sns.barplot(y='Heart Disease Per', x='trestbps', data=data_restbps)
plt.xticks(rotation=90)
plt.title('Disease per bps')
#dont really see a correlation between resting blood pressure and heart disease 
```
